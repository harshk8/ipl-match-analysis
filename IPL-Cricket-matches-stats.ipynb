# Which location hosted most number of matches and win % and loss % for the season


```python
max_played_loc = df_by_season['city'].value_counts().keys()[0]

win_loss_df = df_by_season[df_by_season['city']==max_played_loc][['team1','team2','winner']]
win_loss_df['loser'] = np.where(win_loss_df['team1']==win_loss_df['winner'],win_loss_df['team2'],win_loss_df['team1'])
win_loss_df = win_loss_df[['winner','loser']]
win_in_city = win_loss_df.winner.value_counts().to_dict() #.str.split("t", n = 1, expand = True)  #.to_frame()
loss_in_city = win_loss_df.loser.value_counts().to_dict() #.str.split("t", n = 1, expand = True)  #.to_frame()

win_loss_per = pd.DataFrame(columns=['teams','win(%)','loss(%)'])
win_loss_per['teams'] = np.unique(win_loss_df[['winner','loser']].values)

for key,value in win_in_city.items():
    for i in range(len(win_loss_per['teams'])):
        if win_loss_per.loc[i]['teams']==key:
            win_loss_per.loc[i]['win(%)']= value

for key,value in loss_in_city.items():
    for i in range(len(win_loss_per['teams'])):
        if win_loss_per.loc[i]['teams']==key:
            win_loss_per.loc[i]['loss(%)']= value

win_loss_per['win(%)'].fillna(0, inplace=True)



win_loss_per
```

```python
win_loss_per['win(%)'], win_loss_per['loss(%)']  = win_loss_per['win(%)']*100/ (win_loss_per['win(%)']+win_loss_per['loss(%)']) , win_loss_per['loss(%)']*100/ (win_loss_per['win(%)']+win_loss_per['loss(%)'])
win_loss_per
```
